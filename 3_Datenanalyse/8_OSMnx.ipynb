```python
# import modules
import osmnx as ox

# specify city
place_name = "Eimsb√ºttel, Hamburg, Germany"
# "Munich, Bavaria, Germany"
# "Berlin, 10117, Germany"

# fetch OSM street network from the location
graph = ox.graph_from_place(place_name, network_type='bike')
```

```python
# check the data type of the graph
type(graph)
```

```python
# plot the streets
ox.plot_graph(graph)
```

```python
area = ox.geocode_to_gdf(place_name)
```

```python
#plot
area.plot()
```

```python
# retrieve subway stations
tags = {'amenity': True, 'subway':'subway_station'}
subway = ox.geometries_from_place(place_name, tags)
```

```python
# how many stations
len(subway)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,8))

# plot subway area and stations
area.plot(ax=ax, facecolor= "black")
subway.plot(ax=ax, facecolor= "red", alpha = 0.7)
```
